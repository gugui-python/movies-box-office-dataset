```python
import pandas as pd

# URLs corregidas (asegrate de que existan en tu repo)
url_movies = "https://raw.githubusercontent.com/gugui-python/movies-box-office-dataset/main/enhanced_box_office_data(2000-2024)u.csv"
url_songs = "https://raw.githubusercontent.com/gugui-python/movies-box-office-dataset/main/songs_2000_2020_50k.csv"
url_exports = "https://raw.githubusercontent.com/gugui-python/movies-box-office-dataset/main/34_years_world_export_import_dataset.csv"

# Cargar los datasets
df_movies = pd.read_csv(url_movies)
df_songs = pd.read_csv(url_songs)
df_exports = pd.read_csv(url_exports)

# Mostrar las primeras filas de cada dataset
print("Movies Dataset:")
display(df_movies.head())

print("Songs Dataset:")
display(df_songs.head())

print("Exports Dataset:")
display(df_exports.head())


```

```python
print("Columnas en Movies Dataset:", df_movies.columns)
print("Columnas en Songs Dataset:", df_songs.columns)
print("Columnas en Exports Dataset:", df_exports.columns)

```
