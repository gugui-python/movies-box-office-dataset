# **Actividad pr√°ctica 5.6**

```python
import pandas as pd

# URL del archivo RAW desde tu GitHub
url_movies = "https://raw.githubusercontent.com/gugui-python/movies-box-office-dataset/main/enhanced_box_office_data(2000-2024)u.csv"

# Cargar el dataset de pel√≠culas
df_movies = pd.read_csv(url_movies)

# Verificar las primeras filas
df_movies.head()

```

 ***Gr√°ficos con Matplotlib***

```python
import matplotlib.pyplot as plt

# Agrupamos por g√©nero y calculamos la recaudaci√≥n promedio
genre_group = df_movies.groupby('Genres')['$Worldwide'].mean().sort_values(ascending=False).head(10)

# Creamos el gr√°fico
plt.figure(figsize=(10,6))
genre_group.plot(kind='bar', color='skyblue', grid=True)
plt.title('Top 10 g√©neros con mayor recaudaci√≥n promedio mundial')
plt.ylabel('Recaudaci√≥n promedio ($)')
plt.xlabel('G√©nero')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Este no me gust√≥ por como est√°n agrupados los g√©neros.

```python
# Agrupar por a√±o y sumar la recaudaci√≥n mundial
yearly_revenue = df_movies.groupby('Year')['$Worldwide'].sum()

# Crear gr√°fico de l√≠nea
plt.figure(figsize=(10,6))
plt.plot(yearly_revenue.index, yearly_revenue.values, color='green', marker='o')
plt.title('Recaudaci√≥n mundial total por a√±o')
plt.xlabel('A√±o')
plt.ylabel('Recaudaci√≥n mundial ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Se observa c√≥mo evolucion√≥ la recaudaci√≥n total a lo largo de los a√±os. Desde el 2000 al 2019 aument√≥ much√≠simos, y luego en 2020 ca√≠da coincide con la pandemia. Despu√©s empieza a recuperarse de a poco.

```python
# Promediar los porcentajes
domestic_avg = df_movies['Domestic %'].mean()
foreign_avg = df_movies['Foreign %'].mean()

# Crear gr√°fico de torta
plt.figure(figsize=(6,6))
plt.pie([domestic_avg, foreign_avg], labels=['Dom√©stico', 'Extranjero'], autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Promedio de distribuci√≥n de recaudaci√≥n')
plt.show()
```

Este gr√°fico muestra qu√© proporci√≥n de la recaudaci√≥n proviene de EE.UU. y Canad√° (dom√©stico) y del resto del mundo (extranjero). S√≥lo dos pa√≠ses son un 35%.

```python
# Contar pel√≠culas por pa√≠s de producci√≥n
country_counts = df_movies['Production_Countries'].value_counts().head(10)

# Crear gr√°fico de barras
plt.figure(figsize=(10,6))
country_counts.plot(kind='bar', color='orange', grid=True)
plt.title('Top 10 pa√≠ses productores de pel√≠culas')
plt.xlabel('Pa√≠s')
plt.ylabel('Cantidad de pel√≠culas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Se ven descentrados los √≠ndices del eje X, pero podemos apreciar como USA siempre es el mayor productor (a veces en conjunto con otros pa√≠ses).

***Vamos ahora con Seaborn***

```python
import seaborn as sns
```

```python
# Ver qu√© g√©neros aparecen m√°s
top_genres = df_movies['Genres'].value_counts().head(3).index

# Filtrar solo esos g√©neros
df_top_genres = df_movies[df_movies['Genres'].isin(top_genres)]

# Gr√°fico filtrado
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_top_genres, x='Year', y='$Worldwide', hue='Genres', alpha=0.7)
plt.title('Recaudaci√≥n mundial por a√±o (Top 3 g√©neros)')
plt.xlabel('A√±o')
plt.ylabel('Recaudaci√≥n mundial ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

En este gr√°fico podemos ver que cada puntito es una pel√≠cula, y que a partir del 2019 las pel√≠culas en general recaudaron mucho menos. Adem√°s, las dos pel√≠culas que m√°s recaudaron fueron antes del 2008.

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df_movies, x='Vote_Count', y='$Worldwide', scatter_kws={'alpha':0.5})
plt.title('Relaci√≥n entre cantidad de votos y recaudaci√≥n')
plt.xlabel('Cantidad de votos')
plt.ylabel('Recaudaci√≥n mundial ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

No hay una relaci√≥n tan directa entre la recaudaci√≥n y la cantidad de votos. Lo esperable ser√≠a que las que m√°s recaudan tengan m√°s votos, pero no es as√≠.

```python
# Ver las calificaciones m√°s comunes
top_ratings = df_movies['Rating'].value_counts().head(5).index

# Filtrar solo esas calificaciones
df_top_ratings = df_movies[df_movies['Rating'].isin(top_ratings)]

# Boxplot filtrado
plt.figure(figsize=(12,6))
sns.boxplot(data=df_top_ratings, x='Rating', y='$Worldwide', hue='Rating', palette='pastel', dodge=False)
plt.title('Distribuci√≥n de recaudaci√≥n mundial por rating (Top 5 ratings)')
plt.xlabel('Rating')
plt.ylabel('Recaudaci√≥n mundial ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

Este boxplot muestra la distribuci√≥n de la recaudaci√≥n mundial ($Worldwide) de las pel√≠culas seg√∫n su calificaci√≥n (Rating). Adem√°s, filtramos por las 5 calificaciones m√°s frecuentes.

**Interpretaci√≥n:**
Caja central: Representa el 50% central de los datos (del 25% al 75%). Esto nos dice la recaudaci√≥n t√≠pica de las pel√≠culas con esa calificaci√≥n.

L√≠neas de los bigotes: Muestran los valores m√≠nimos y m√°ximos, a excepci√≥n de los valores at√≠picos.

Outliers (puntos fuera de los bigotes): Son las pel√≠culas que recaudaron mucho m√°s o mucho menos que las dem√°s. Esto puede indicar grandes √©xitos o fracasos.

Entonces:
- Una pel√≠cula con rating 6.3 recaud√≥ m√°s que otras con m√°s rating.
- La caja verde (rating 6.5) tuvo valores muy altos (l√≠nea bigote de arriba).
- No hay outliers m√°s bajos que la media.
